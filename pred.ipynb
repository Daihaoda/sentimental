```python
import pandas as pd


# Loading news data and stock data
news_data = pd.read_csv('Combined_News_DJIA.csv')
stock_data = pd.read_csv('DJIA_table.csv')


# View basic data information
print("News Data：")
display(news_data.head())
print("Stock Data：")
display(stock_data.head())



```

```python
# 检查日期范围和数量
print("Stock data date range：", stock_data['Date'].min(), "to", stock_data['Date'].max())
print("News data date range：", news_data['Date'].min(), "to", news_data['Date'].max())
print("Number of rows of stock data：", len(stock_data))
print("Number of rows of news data：", len(news_data))

# View the intersection of dates in two data frames
common_dates = set(stock_data['Date']).intersection(set(news_data['Date']))
print("Number of shared dates：", len(common_dates))

```

```python
# Convert the date format to YYYY-MM-DD
stock_data['Date'] = pd.to_datetime(stock_data['Date'], format='%d-%m-%Y').dt.strftime('%Y-%m-%d')
news_data['Date'] = pd.to_datetime(news_data['Date']).dt.strftime('%Y-%m-%d')

# Check the converted date
print("Date format of converted stock data：")
print(stock_data['Date'].head())
print("Date format of converted news data：")
print(news_data['Date'].head())
```

```python
# Check the number of shared dates
common_dates = set(stock_data['Date']).intersection(set(news_data['Date']))
print("Number of shared dates：", len(common_dates))

# Combining data
data = pd.merge(stock_data, news_data, on='Date', how='inner')
data.fillna('', inplace=True)

# Checking the merged data
display(data.head())

```

```python
# Combine news headlines into a single text column
data['Combined_News'] = data[[f'Top{i}' for i in range(1, 26)]].apply(lambda x: ' '.join(x), axis=1)

# View merged news text
data[['Date', 'Combined_News']].head()

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Merge all news headlines into one string
all_news_text = ' '.join(data['Combined_News'])

# Customized stop words (can be added as needed)
custom_stopwords = set(STOPWORDS)

# Generate word clouds and remove common nonsense words
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=custom_stopwords).generate(all_news_text)

# Show Word Cloud Map
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('News Headlines Word Cloud')
plt.show()


```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Extracting news text features using TF-IDF and removing common nonsense words
tfidf = TfidfVectorizer(max_features=1000, stop_words='english')
news_features = tfidf.fit_transform(data['Combined_News']).toarray()

# Check the dimension of the extracted features
print("News Feature Dimension:", news_features.shape)

# Get feature name and corresponding frequency
tfidf_feature_names = tfidf.get_feature_names_out()
tfidf_feature_weights = np.asarray(news_features.sum(axis=0)).flatten()

# Building and Sorting DataFrames
tfidf_df = pd.DataFrame({'feature': tfidf_feature_names, 'weight': tfidf_feature_weights})
tfidf_df = tfidf_df.sort_values(by='weight', ascending=False).head(20)  # Showing the first 20 features

# Plotting bar charts
plt.figure(figsize=(12, 6))
plt.barh(tfidf_df['feature'], tfidf_df['weight'])
plt.gca().invert_yaxis()
plt.title('Top 20 TF-IDF Features')
plt.xlabel('Weight')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

def create_dataset(features, target, time_step):
    x, y = [], []
    n_vectors = int(len(features) / (time_step + 1))
    for n in range(n_vectors):
        # Create time-step windows for each feature
        feature_windows = [features[n * (time_step + 1):n * (time_step + 1) + time_step, col] for col in range(features.shape[1])]
        combined_features = np.column_stack(feature_windows)
        x.append(combined_features)
        y.append(target[n * (time_step + 1) + time_step])
    return np.array(x), np.array(y)

# Defining Time Steps
time_step = 5  # For example, 5 time steps

# Creating a dataset with the create_dataset function
X, y = create_dataset(features, target, time_step)

# Chronological division of training and test sets
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print("Training Set Size:", X_train.shape)
print("Test Set Size:", X_test.shape)

# Plot the distribution of target values for the training and test sets
plt.figure(figsize=(12, 6))
plt.hist(y_train, bins=30, alpha=0.5, label='Training Set')
plt.hist(y_test, bins=30, alpha=0.5, label='Testing Set')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Distribution of Closing Prices in Training and Testing Sets')
plt.legend()
plt.show()



```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Flatten X_test to 2D
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# 进行预测
y_pred = model.predict(X_test_flat)  # Multiply by 105 if needed, can adjust it directly here

# Calculation of assessment indicators
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse}")
print(f"MAE: {mae}")
print(f"R²: {r2}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Retain only stock features
features = data[['Open', 'High', 'Low', 'Volume']].values  # Using stock features instead of news + stock features
target = data['Close'].values

# The create dataset function remains the same
def create_dataset(features, target, time_step):
    x, y = [], []
    n_vectors = int(len(features) / (time_step + 1))
    for n in range(n_vectors):
        feature_windows = [features[n * (time_step + 1):n * (time_step + 1) + time_step, col] for col in range(features.shape[1])]
        combined_features = np.column_stack(feature_windows)
        x.append(combined_features)
        y.append(target[n * (time_step + 1) + time_step])
    return np.array(x), np.array(y)

# Defining Time Steps
time_step = 5  # For example, 5 time steps

# Creating a dataset using stock features
X, y = create_dataset(features, target, time_step)

# Chronological division of training and test sets
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Flatten X_train and X_test to two dimensions for linear regression modeling.
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# Training a linear regression model
model = LinearRegression()
model.fit(X_train_flat, y_train)
print("Model training completed")

# executing forecast
y_pred = model.predict(X_test_flat)

# Calculation of assessment indicators
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MSE: {mse}")
print(f"MAE: {mae}")
print(f"R²: {r2}")

# Plotting line graphs of predicted and true values
plt.figure(figsize=(14, 7))
plt.plot(y_test , label='True', color='blue')
plt.plot(y_pred, label='Pred', color='orange')
plt.title('DJIA Close Price Prediction Without News Data')
plt.xlabel('Sample')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
